```python
#  YalnÄ±zca analizimle alakalÄ± sÃ¼tunlarÄ± seÃ§tim

df = df[['Artist', 
    #'Url_spotify', 
    'Track', 'Album', 'Album_type', 
    #'Uri',
    'Danceability', 
    #'Energy', 'Key', 'Loudness', 
    'Speechiness', 
    #'Acousticness', 'Instrumentalness', 'Liveness', 
    'Valence', 
    #'Tempo', 
    'Duration_ms', 
    #'Url_youtube', 
    'Title', 'Channel', 'Views', 'Likes', 'Comments', 
    #'Description', 'Licensed', 
    'official_video', 'Stream']].copy()
```

```python
# SÃ¼tunlarÄ±mÄ± yeniden adlandÄ±rÄ±yorum.

df = df.rename(columns={'Album_type':'Album_Tipi',
                      'Title':'Youtube_BaslÄ±klar',
                      'Channel':'Youtube_KanallarÄ±',
                      'Views':'Youtube_GÃ¶rÃ¼ntÃ¼leme',
                      'Likes':'Youtube_Begeniler',
                      'Comments':'Youtube_Yorumlar',
                      'official_video':'Official_Vidyo',
                      'Stream':'Youtube_AkÄ±s'})
```

```python
# bu dÃ¶nÃ¼ÅŸÃ¼mÃ¼ veri analizi ve gÃ¶rselleÅŸtirme iÃ§in daha uygun bir formatta olsun dÄ±ye yaptÄ±m.

# ÅŸarkÄ±nÄ±n dakika cinsinden sÃ¼resini iÃ§eren yeni sÃ¼tun ekledim

df['Duration_Minutes'] = df['Duration_ms'] / 60000
```

```python
# null verileri kontrol ettim.

df.isna().sum()
```

```python
#Her bir sÃ¼tundaki boÅŸ deÄŸerleri kaldÄ±rmadan Ã¶nce belirli bir 
#eÅŸik deÄŸerini kontrol ediyorum. EÄŸer bir sÃ¼tundaki boÅŸ deÄŸerler, 
#toplam satÄ±r sayÄ±sÄ±nÄ±n %5'inden azsa, 
#yani o sÃ¼tundaki boÅŸ deÄŸerlerin oranÄ± belirli bir eÅŸiÄŸi geÃ§miyorsa,
#o sÃ¼tundaki boÅŸ deÄŸerler kaldÄ±rÄ±lacak.

df.dropna(subset=df.columns[df.isna().sum() <= len(df) * 0.05], 
          inplace=True)

```

```python
#yinelenen deÄŸerleri kaldÄ±rarak sÄ±fÄ±rlÄ±yorum.

df = df.drop_duplicates() \
    .reset_index(drop=True).copy()
```

```python
# veri tÃ¼rlerini dÃ¶nÃ¼ÅŸtÃ¼rÃ¼yorum.

kolon_donustur = ['Youtube_GÃ¶rÃ¼ntÃ¼leme', 'Youtube_Begeniler', 'Youtube_Yorumlar', 'Youtube_AkÄ±s']

for col in kolon_donustur:
    df[col] = df[col].astype('int')

df['Official_Vidyo'] = df['Official_Vidyo'].astype('str')
df['Official_Vidyo'] = df['Official_Vidyo'].str.replace('True', 'Yes')
df['Official_Vidyo'] = df['Official_Vidyo'].str.replace('False', 'No')
```

```python
df_grouped = df.groupby('Artist')[['Youtube_GÃ¶rÃ¼ntÃ¼leme', 'Youtube_AkÄ±s']].sum().reset_index()
```

```python
df_grouped = df_grouped.sort_values(by='Youtube_GÃ¶rÃ¼ntÃ¼leme', ascending=False)[:10]
```

```python
df_melted = pd.melt(df_grouped, id_vars=['Artist'], var_name='Platform', value_name='Counts')
```

```python
sns.barplot(y='Artist', x='Counts', hue='Platform', data=df_melted, orient = 'h')
plt.xticks(rotation=0)
plt.xlabel('Artist')
plt.ylabel('Counts')
plt.title('YouTube ve Spotifydaki En Ä°yi 10 SanatÃ§Ä±');
```
